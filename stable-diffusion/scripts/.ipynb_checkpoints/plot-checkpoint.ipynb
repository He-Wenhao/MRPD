```python
import matplotlib.pyplot as plt

# Initialize lists
psnr_list = []
ssim_list = []
mse_list = []

# Read the file and parse values
with open('intermediate.txt', 'r') as f:
    lines = f.readlines()[1:]  # skip header line

    for line in lines:
        parts = line.strip().split(',')
        psnr = float(parts[0].split(':')[1])
        ssim = float(parts[1].split(':')[1])
        mse = float(parts[2].split(':')[1])
        psnr_list.append(psnr)
        ssim_list.append(ssim)
        mse_list.append(mse)

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

axs[0].plot(psnr_list, marker='o', markersize=2)
axs[0].set_ylabel('PSNR')
axs[0].set_title('PSNR over Samples')

axs[1].plot(ssim_list, marker='o', markersize=2)
axs[1].set_ylabel('SSIM')
axs[1].set_title('SSIM over Samples')

axs[2].plot(mse_list, marker='o', markersize=2)
axs[2].set_ylabel('MSE')
axs[2].set_title('MSE over Samples')
axs[2].set_xlabel('Sample Index')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset

# Initialize lists
psnr_list = []
ssim_list = []
mse_list = []

# Read the file and parse values
with open('intermediate.txt', 'r') as f:
    lines = f.readlines()[1:]  # skip header line
    for line in lines:
        parts = line.strip().split(',')
        psnr = float(parts[0].split(':')[1])
        ssim = float(parts[1].split(':')[1])
        mse = float(parts[2].split(':')[1])
        psnr_list.append(psnr)
        ssim_list.append(ssim)
        mse_list.append(mse)

# Shift x-axis to 6001200
x_main = list(range(600, 600 + len(psnr_list)))
x_inset_start = 1000
x_inset_end = 1200
i_start = x_inset_start - 600
i_end = x_inset_end - 600

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

metric_lists = [psnr_list, ssim_list, mse_list]
titles = ['PSNR over Samples', 'SSIM over Samples', 'MSE over Samples']
ylabels = ['PSNR', 'SSIM', 'MSE']

for i in range(3):
    axs[i].plot(x_main, metric_lists[i], marker='o', markersize=2)
    axs[i].axvline(x=1000, color='red', linestyle='--', linewidth=1)  # Red dashed line
    axs[i].set_ylabel(ylabels[i])
    axs[i].set_title(titles[i])

    # Add inset
    inset_ax = inset_axes(axs[i], width="35%", height="35%", loc='center right')
    inset_ax.plot(x_main, metric_lists[i], marker='o', markersize=1)
    inset_ax.set_xlim(x_inset_start, x_inset_end)
    y_window = metric_lists[i][i_start:i_end + 1]
    y_min, y_max = min(y_window), max(y_window)
    inset_ax.set_ylim(y_min, y_max)
    inset_ax.tick_params(axis='both', labelsize=8)
    inset_ax.set_title("Zoom (10001200)", fontsize=8)

    mark_inset(axs[i], inset_ax, loc1=2, loc2=4, fc="none", ec="0.5")

axs[2].set_xlabel('Sample Index')
plt.tight_layout()
plt.savefig("report/loss.png")
```

```python
from PIL import Image, ImageDraw

# Configuration
num_images = 20
grid_cols = 3
grid_rows = 6
image_prefix = "results/comparison_"
image_suffix = ".png"
output_filename = "assembled.png"

# Load all images
images = [Image.open(f"{image_prefix}{i}{image_suffix}") for i in range(1, num_images + 1)]

# Get size of individual images (assumes all same size)
img_width, img_height = images[0].size

# Create a blank canvas for the assembled image
canvas_width = grid_cols * img_width
canvas_height = grid_rows * img_height
assembled_img = Image.new('RGB', (canvas_width, canvas_height), color='white')

# Paste images into grid
for idx, img in enumerate(images):
    row = idx // grid_cols
    col = idx % grid_cols
    if row < grid_rows:
        assembled_img.paste(img, (col * img_width, row * img_height))

# Draw vertical column separators
draw = ImageDraw.Draw(assembled_img)
for col in range(1, grid_cols):
    x = col * img_width
    draw.line([(x, 0), (x, canvas_height)], fill='black', width=2)

# Save the final image
assembled_img.save(output_filename)
print(f"Saved assembled image as {output_filename}")
```
